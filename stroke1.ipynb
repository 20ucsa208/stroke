```python
import pandas as pd
cr=pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
cr.info()
```

```python
cr.isnull().sum()
cr.shape
```

```python
cr["bmi"].fillna(cr["bmi"].mean(),inplace=True)
```

```python
cr.isnull().sum()
```

```python
cr.drop("id",axis=1,inplace=True)
cr
```

```python
cr.gender.replace({"Male":0,"Female":1,"Other":2},inplace=True)
cr.ever_married.replace({"Yes":0,"No":1},inplace=True)
cr.work_type.replace({"Private":0,"Self-employed":1,"Govt_job":2,"children":3,"Never_worked":4},inplace=True)
cr.Residence_type.replace({"Urban":0,"Rural":1},inplace=True)
cr.smoking_status.replace({"formerly smoked":0,"never smoked":1,"smokes":2,"Unknown":3},inplace=True)
cr
```

```python
X=cr.iloc[:,:-1]
Y=cr.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
X_train_std=std.fit_transform(X_train)
X_test_std=std.transform(X_test)
```

```python
import pickle
import os
scaler_path=os.path.join("C:/Users/User/stroke/",'models/scaler.pkl')
with open(scaler_path,'wb') as scaler_file:
    pickle.dump(std,scaler_file)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train_std,Y_train)
Y_pred=dt.predict(X_test_std)
from sklearn.metrics import accuracy_score
ac_dt=accuracy_score(Y_test,Y_pred)
ac_dt
```

```python
import joblib
model_path=os.path.join('C:/Users/User/stroke/','models/dt.sav')
joblib.dump(dt,model_path)
```
