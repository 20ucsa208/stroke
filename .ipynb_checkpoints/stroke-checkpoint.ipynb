```python
import pandas as pd
cr=pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cr.info()
```

```python
cr.isnull().sum()
```

```python
cr.shape
```

```python
cr["bmi"].fillna(cr["bmi"].mean(),inplace=True)
```

```python
cr.isnull().sum()
```

```python
cr.drop("id",axis=1,inplace=True)
cr
```

```python
cr.gender.replace({"Male":0,"Female":1,"Other":2},inplace=True)
cr.ever_married.replace({"Yes":0,"No":1},inplace=True)
cr.work_type.replace({"Private":0,"Self-employed":1,"Govt_job":2,"children":3,"Never_worked":4},inplace=True)
cr.Residence_type.replace({"Urban":0,"Rural":1},inplace=True)
cr.smoking_status.replace({"formerly smoked":0,"never smoked":1,"smokes":2,"Unknown":3},inplace=True)
```

```python
cr
```

```python
cr_x=cr.iloc[:,:-1]
cr_y=cr.iloc[:,-1]
```

```python
cr_x.head()
```

```python
cr_y.head()
```

```python
from sklearn.model_selection import train_test_split
cr_x_train,cr_x_test,cr_y_train,cr_y_test=train_test_split(cr_x,cr_y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=100)
rfc.fit(cr_x_train,cr_y_train)
pred=rfc.predict(cr_x_test)
from sklearn.metrics import accuracy_score
accuracy_score(cr_y_test,pred)*100
```

```python
from sklearn.metrics import confusion_matrix
result=confusion_matrix(cr_y_test,pred)
rfc_result=result.diagonal().sum()/result.sum()*100
rfc_result
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(cr_x_train,cr_y_train)
Y_pred_lr=lr.predict(cr_x_test)
ac_lr=accuracy_score(cr_y_test,Y_pred_lr)
ac_lr
import pickle
filename=r"C:\Users\User\stroke\finalized_model.sav"
pickle.dump(lr,open(filename,"wb"))
```
